
Select all the even indexed rows from the given dataset. What is the value in the 3rd indexed column of the 0th indexed row in the selected dataframe. (row/ column indexed from zero(0) in python/pandas)


```python
import pandas as pd
df = pd.read_csv('/content/V5.csv')
# Replace "?" with NaN
df.replace('?', float('nan'), inplace=True)
even_indexed_rows = df.iloc[::2]
value = even_indexed_rows.iloc[0, 3]
print("Value in the 3rd indexed column of the 0th indexed row in the selected dataframe:", value)
```

Determine the number of samples (rows) in the dataset that belong to the six most recent years, based on the provided data.


```python
import pandas as pd
df = pd.read_csv('/content/V5.csv')
df.replace('?', float('nan'), inplace=True)
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
most_recent_year = pd.to_datetime('now').year
six_most_recent_years = list(range(most_recent_year, most_recent_year - 6, -1))
recent_samples = df[df['Year'].dt.year.isin(six_most_recent_years)]
num_recent_samples = recent_samples.shape[0]
print("Number of samples from the six most recent years:", num_recent_samples)
print("Recent Years:",six_most_recent_years)
```

```python
import pandas as pd
df = pd.read_csv('/content/V5.csv')
df.replace('?', float('nan'), inplace=True)
most_recent_years = df['Year'].max()
six_most_recent_years = range(most_recent_years - 5, most_recent_years + 1)
recent_samples = df[df['Year'].isin(six_most_recent_years)]
num_recent_samples = recent_samples.shape[0]
print("Number of samples from the six most recent years based on the data:", num_recent_samples)
```

Determine the number of samples (rows) in the dataset which has num_rooms = 3 and num_bathrooms = 3

```python
import pandas as pd
df = pd.read_csv('/content/V5.csv')
df.replace('?', float('nan'), inplace=True)
filtered_data = df[(df['num_rooms'] == 3) & (df['num_bathrooms'] == 3)]
num_matching_samples = filtered_data.shape[0]
print("Number of samples with num_rooms = 3 and num_bathrooms = 3:", num_matching_samples)
```

In the Year 2022, how many houses (rows) located in the Greenwich Locality have exactly num_room == 3, and facing towards either the North or East?

```python
import pandas as pd
df = pd.read_csv('/content/V5.csv')
df.replace('?', float('nan'), inplace=True)
filtered_data = df[(df['Year'] == 2022) &
                   (df['Locality'] == 'Greenwich') &
                   (df['num_rooms'] == 3) &
                   ((df['Face'] == 'North') | (df['Face'] == 'East'))]
num_matching_samples = filtered_data.shape[0]
print("Number of houses in Greenwich in the year 2022 with num_rooms = 3 and facing North or East:", num_matching_samples)
```

Calculate the number of samples(rows) that correspond to the month of August across all the years.?

```python
import pandas as pd
df = pd.read_csv('/content/V5.csv')
df.replace('?', float('nan'), inplace=True)
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
august_samples = df[df['Date'].dt.month == 8]
num_august_samples = august_samples.shape[0]
print("Number of samples corresponding to the month of August across all years:", num_august_samples)
```

According to the data which Locality have highest average Sale Price?

```python
import pandas as pd
df = pd.read_csv('/content/V5.csv')
df.replace('?', float('nan'), inplace=True)
locality_avg_sale_price = df.groupby('Locality')['Sale Price'].mean()
highest_avg_price_locality = locality_avg_sale_price.idxmax()
highest_avg_price = locality_avg_sale_price.max()
print(f"The locality with the highest average sale price is '{highest_avg_price_locality}' with an average sale price of ${highest_avg_price:.2f}")
```
