How many benign ( B ) tumour cases are present in the Wisconsin breast cancer dataset ?

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
target_labels = data.target
num_benign_cases = (target_labels == 0).sum()
print("Number of benign (B) tumor cases:", num_benign_cases)
```

How many malignant ( M ) tumour cases are present in the Wisconsin breast cancer dataset.


```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
target_labels = data.target
num_malign_cases = (target_labels == 1).sum()
print("Number of malignant (M) tumor cases:", num_malign_cases)
```
